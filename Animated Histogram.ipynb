```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)

plt.figure(figsize=(9,3))
plt.hist(x1, density=True, bins=20, alpha=0.5)
plt.hist(x2, density=True, bins=20, alpha=0.5)
plt.axis([-7,15,0,0.6])

_ = plt.text(x1.mean()-1.5, 0.5, 'x1\nNormal')
_ = plt.text(x2.mean()-1.5, 0.5, 'x2\nGamma')
```

```python
import matplotlib.animation as animation

f=250

def update_normal(curr):
    if curr==f:
        n.event_source.stop()

    plt.cla()
    bins = np.arange(-6.5, 1.5, 0.5)
    plt.hist(x1[:curr], bins=bins, alpha=0.5, density=True)
    plt.axis([-6.5,1.5,0,0.6])
    plt.annotate('n = {}'.format(curr), [0,0.4])
    plt.gca().set_title('Normal Dist. Animation')
    
def update_gamma(curr):
    if curr==f:
        g.event_source.stop()

    plt.cla()
    bins = np.arange(-1, 15, 0.5)
    plt.hist(x2[:curr], bins=bins, alpha=0.5, density=True)
    plt.axis([-1,15,0,0.43])
    plt.annotate('n = {}'.format(curr), [13,0.4])
    plt.gca().set_title('Gamma Dist. Animation')
```

```python
normal = plt.figure()
n = animation.FuncAnimation(normal, update_normal, interval=5)
```

```python
gamma = plt.figure()
g = animation.FuncAnimation(gamma, update_gamma, interval=5)
```
